```python
contacts = {}

def insert_contact():
    name = input("Enter the contact name: ").strip().lower()
    while True:
        phone = input("Enter the contact phone number (11 digits): ").strip()
        if phone.isdigit() and len(phone) == 11:
            contacts[name] = phone
            print(f"Contact {name} with phone number {phone} has been added successfully.")
            break
        else:
            print("Invalid phone number. Please enter a numeric 11 digit phone number")
            
def read_contacts():
    if not contacts:
        print("No contacts found.")
    else:
        print("Contacts:")
        for name, phone in contacts.items():
            print(f"{name.capitalize()}: {phone}")
            
def update_contacts():
    name = input("Enter the contact name to update: ").strip().lower()
    if name in contacts:
        while True:
            phone = input("Enter the new contact phone number (11 digits): ").strip()
            if phone.isdigit() and len(phone) == 11:
                contacts[name] = phone
                print(f"Contact {name} with phone number {phone} has been added successfully.")
                break
            else:
                print("Invalid phone number. Please enter a numeric 11 digit phone number")
    else:
        print(f"Contact {name} not found.")
        
def delete_contact():
    name = input("Enter the contact name to delete: ").strip().lower()
    if name in contacts:
        del contacts[name]
        print(f"Contact {name} has been deleted successfully.")
    else:
        print(f"Contact {name} not found.")
        
print("Welcome to the contacts app.")
print("*" * 30) 
print()
while True:
    print("Choose an option:")
    print("1. Insert contact")
    print("2. Read contacts")
    print("3. Update contact")
    print("4. Delete contact")
    print("5. Exit")
    try:
        option = int(input("Enter your option: "))
        if option == 1:
            insert_contact()
        elif option == 2:
            read_contacts()
        elif option == 3:
            update_contacts()
        elif option == 4:
            delete_contact()
        elif option == 5:
            print("Goodbye!")
            break
        else:
            print("Invalid option. Please enter a valid option.")
    except ValueError:
        print("Invalid option. Please enter a numeric value between 1 and 5.")        
            
```
